```python
df['ValuePercentage'] = df['ValuePercentage'].astype('int')
df['Year'] = df['Year'].astype('category')
df['Measure'] = df['Measure'].astype('category')
```

```python
df.sort_values(by='ValuePercentage', inplace=True, ascending=True)
df.tail(5)
```

```python
df['ValuePercentage'].unique
```

```python
df.info()
```

```python
df['MeasureList'] = np.select(
[
    ( df['Measure'] == 'Percentoftotaldeaths' )
], 
[
    1
], 
default=2 )

df['MeasureList']
```

```python
df_causes_death = df.query(" MeasureList == 1")
df_change_percent = df.query(" MeasureList == 2")
```

```python
df_causes_death.count()
```

```python
df_change_percent.count()
```

```python
df_causes_death['ValuePercentage']
```

```python
df_change_percent['ValuePercentage']
```

```python
df_otros_death = df_causes_death.query("((ValuePercentage >= 0) & (ValuePercentage < 1))")
df_otros_death.shape
```

```python
import numpy as np

df_causes_death['Measure'] = np.select(
[
    ( df_causes_death['CauseDeathOrInjury'].isin(df_otros_death["CauseDeathOrInjury"].to_list()))
], 
[
    'Otros'
], 
default = df_causes_death['Measure'] )

df_causes_death['Measure']
```

## Dataset medida 2017 . % Deaths 

```python
df_causes_death_otros = df_causes_death.query('Measure == "Otros"')
df_causes_death_otros
```

```python
df_causes_death_otros = df_causes_death_otros.groupby(['Measure']).sum()
df_causes_death_otros
```

```python
df_causes_death_otros['CauseDeathOrInjury'] = 'Otros'
df_causes_death_otros['Year'] = '2017'
df_causes_death_otros['Measure'] = 'Otros'
df_causes_death_otros['CaseDeathList'] = 2
df_causes_death_otros['MeasureList'] = 2

df_causes_death_otros.tail(5)
```

```python
df_causes_death = df_causes_death[df_causes_death.Measure != "Otros"]
```

```python
frames = [df_causes_death, df_causes_death_otros]
df_causes_death = pd.concat(frames)
```

```python
df_causes_death.shape
```

```python
df_causes_death
```

```python
df_causes_death.to_excel('Medida1_2017.xlsx')
```

## Dataset medida 2005 .  Change % Deaths 

```python
df_change_percent['Measure'] = np.select(
[
    ( df_change_percent['CauseDeathOrInjury'].isin(df_otros_death["CauseDeathOrInjury"].to_list()))
], 
[
    'Otros'
], 
default = df_change_percent['Measure'] )

df_change_percent['Measure']
```

```python
df_change_percent.query('Measure == "Otros"').count()
```

```python
df_change_percent_otros = df_change_percent.query('Measure == "Otros"')
```

```python
df_change_percent_otros = df_change_percent_otros.groupby(['Measure']).sum()
df_change_percent_otros
```

```python
df_change_percent_otros['CauseDeathOrInjury'] = 'Otros'
df_change_percent_otros['Year'] = '2005 - 2010'
df_change_percent_otros['Measure'] = 'Otros'
df_change_percent_otros['CaseDeathList'] = 1
df_change_percent_otros['MeasureList'] = 1

df_change_percent_otros.tail(5)
```

```python
df_change_percent = df_change_percent[df_change_percent.Measure != "Otros"]
```

```python
frames = [df_change_percent, df_change_percent_otros]
df_change_percent = pd.concat(frames)
```

```python
df_change_percent.shape
```

```python
df_change_percent.to_excel('Medida1_2005.xlsx')
```

## Pivot DS 2017

```python
df_causes_death = df_causes_death.pivot(index=['CauseDeathOrInjury',
                                                'Year',                                                
                                                'Value',                                                
                                                'ValuePercentage',
                                                'CaseDeathList',
                                                'Percent',
                                                'MeasureList'], 
                                        columns=['Measure'], 
                                        values='Percent')

df_causes_death
```

```python
df_causes_death['Otros'] =  df_causes_death['Otros'].fillna(0)
df_causes_death['Percentoftotaldeaths'] = df_causes_death['Percentoftotaldeaths'].fillna(0)
df_causes_death['Deathsannualchange'] = 0
```

```python
df_causes_death
```

## Pivot DS 2005

```python
df_change_percent = df_change_percent.pivot(index=['CauseDeathOrInjury',
                                                'Year',                                                
                                                'Value',                                                
                                                'ValuePercentage',
                                                'CaseDeathList',
                                                'Percent',
                                                'MeasureList'], 
                                        columns=['Measure'], 
                                        values='Percent',)

df_change_percent
```

```python
df_change_percent['Otros'] = df_change_percent['Otros'].fillna(0)
df_change_percent['Percentoftotaldeaths'] = 0
df_change_percent['Deathsannualchange'] = df_change_percent['Deathsannualchange'].fillna(0)
```

```python
df_change_percent
```

## Union

```python
df_change_percent = df_change_percent.sort_values(by=['Deathsannualchange'], ascending=False)
df_change_percent
```

```python
df_causes_death = df_causes_death.sort_values(by=['Percentoftotaldeaths'], ascending=False)
df_causes_death
```

```python
frames = [
          df_causes_death, 
          df_change_percent
         ]
result = pd.concat(frames)

result['orden'] = np.arange(len(result))

result.shape
```
